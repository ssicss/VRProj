# Foculus Rift data dumper

In this notebook I use the pyUSB library to connect to the STM32F3DISCOVERY board, running the Foculus Rift firmware. I set up the scale and sample rate settings and record data for about 1 minute. THe data is decoded as float values and saved to disk as .npz file. The second half of the notebook plots these .npz files.

YFL, 2014

```python
import sys
import usb.core
import struct
import time
from mpl_toolkits.mplot3d import Axes3D, proj3d
#PRODUCT_ID = 0x5710
PRODUCT_ID = 0x0001
#HID requests:
GET_REPORT = 1
SET_REPORT = 9
N = 30000  #Number of packets to read

def printHex( byteBlock ):
    """ display a block of hex numbers for debug purposes """
    sys.stdout.write("\n          0x00 0x01 0x02 0x03 0x04 0x05 0x06 0x07 0x08 0x09 0x0A 0x0B 0x0C 0x0D 0x0E 0x0F")
    sys.stdout.write("\n         --------------------------------------------------------------------------------")
    sys.stdout.write("\n0x000000| ")
    for i,b in enumerate( byteBlock ):
        sys.stdout.write( "0x%02X "%b )
        if (i+1)%16 == 0:
            sys.stdout.write("\n0x%06X| "%(i+1) )
    sys.stdout.write("\n")
```

# Step1, connect to device

```python
d = usb.core.find( idProduct = PRODUCT_ID )
if (d.is_kernel_driver_active(0)):
    d.detach_kernel_driver(0)
    print "Kernel driver detached"
d.set_configuration()
if d is not None:
    print "Connected!"
```

# Setup scale and interval for 1 ms

```python
#//static const UInt16 AccelRangeRamp[] = { 2, 4, 8, 16 };
#//static const UInt16 GyroRangeRamp[]  = { 250, 500, 1000, 2000 };
#//static const UInt16 MagRangeRamp[]   = { 880, 1300, 1900, 2500 };
#------------------------------------
accscale = 8
gyroscale = 250
magnetoscale = 2500
sendIntervalMs = 1
#------------------------------------
datSetScale  = array([0x04, 0x00, 0x00, accscale&0xFF, gyroscale&0xFF, gyroscale>>8, magnetoscale&0xFF, magnetoscale>>8], dtype=uint8)
datSetConfig = array([0x02, 0x00, 0x00, 0x00, sendIntervalMs&0xFF, 0xFF, 0xFF], dtype=uint8)
printHex( datSetScale )
printHex( datSetConfig )
print d.ctrl_transfer( 0x21, SET_REPORT, 0x0304, 0, datSetScale)
print d.ctrl_transfer( 0x21, SET_REPORT, 0x0302, 0, datSetConfig)   #Send config frame, request 10 ms rate
dat = d.ctrl_transfer( bmRequestType=0xA1, bRequest=GET_REPORT, wValue=0x0302, wIndex=0, data_or_wLength=100)
printHex( dat )
```

# Collect some sensor data and save to disk

```python
FNAME = "afterCalib.npz"
d.ctrl_transfer( 0x21, SET_REPORT, 0x0308, 0, [0x08, 0x00, 0x00, 0xFF, 0xFF])   #send keep alive
readRawData = zeros((N,62), dtype=uint8)
startTime = time.time()
for n in range(N):
    readRawData[n,:] = d.read(0x81, 64, timeout=1000)
stopTime = time.time()
#Save to disk
savez(FNAME, readRawData=readRawData)
print "Recorded data for %.3f s to %s"%(stopTime-startTime, FNAME)
```

# Load results form disk and decode the results

```python
def unpackReading( rawSensorBlock ):
    """
    The Rift IMU sensor values are 21 bit signed integers, packed in a datablock in a rather random order, here we unpack them
    In:     8 byte raw datablock, containing X, Y, Z sensor values of Gyro or Acc
    Out:    array of signed 32 bit integers, containing the 21 bit signed integers
    Remark: Max range corresponds to:   - 1 048 576   ...   + 1 048 575
    """
    readings = zeros(3, dtype=int32)
    readings[0] = int32( rawSensorBlock[0]<<13  | rawSensorBlock[1]<<5 | rawSensorBlock[2]>>3 )
    readings[1] = int32( (rawSensorBlock[2]&0x07)<<18 | rawSensorBlock[3]<<10 | rawSensorBlock[4]<<2 | rawSensorBlock[5]>>6 )
    readings[2] = int32( (rawSensorBlock[5]&0x3F)<<15 | rawSensorBlock[6]<<7 | rawSensorBlock[7]>>1 )
    for i in range(3):
        #Check if sign extension is needed
        if readings[i] >   0x00FFFFF:   #We have a negative number! Extend sign! (twos complement)
            readings[i] -= 0x0200000
    return readings
```

```python
#FNAME = "sensordat.npz"
readRawData = load(FNAME)["readRawData"]
gyroData = zeros((N,3), dtype=float)
accData = zeros((N,3), dtype=float)
magData = zeros((N,3), dtype=float)
for n in xrange(N):
    unpackedData  = struct.unpack('<BBHHh16B16B16Bhhh', readRawData[n,:])
    magData[n,:]  = array(unpackedData[-3:], dtype=float)   * 1E-4 #[G]
    accData[n,:]  = unpackReading( unpackedData[5:21] )     * 1E-4 #[m/s^2]
    gyroData[n,:] = unpackReading( unpackedData[8+5:8+21] ) * 1E-4 #[rad/s]
```
